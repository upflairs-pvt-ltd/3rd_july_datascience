### pie plot 

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']  



plt.pie(country_population,labels=country_name)
plt.show()
```

```python
plt.pie(country_population,labels=country_name,autopct="%2.2f%%")
plt.show()
```

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']  
explosion = [0.2,0,0,0,0.4]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",explode=explosion)
plt.show()
```

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']  
explosion = [0.2,0,0,0,0.4]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",explode=explosion,wedgeprops={'ec':'black','lw':3})
plt.show()
```

```python
plt.figure(figsize=(6,8))   # widht and height 
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']  
explosion = [0.2,0,0,0,0.4]
plt.pie(country_population,labels=country_name,autopct="%2.2f%%",explode=explosion,wedgeprops={'ec':'black','lw':3},shadow=True)
plt.savefig('country_pipplot')
plt.show()
```
