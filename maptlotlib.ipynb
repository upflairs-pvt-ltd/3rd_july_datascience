```python
df = pd.read_csv('Used_Bikes.csv')
df.head() 
```

```python
df.drop_duplicates(inplace=True)   # 25k 
```

```python
brand_name = list(df['brand'].value_counts().keys())
brand_name
```

```python
no_of_count = list(df['brand'].value_counts().values)
no_of_count
```

```python
brand_name = list(df['brand'].value_counts().keys())
no_of_count = list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count)
plt.xticks(rotation=75) 
# plt.yticks(rotation=75)

plt.show()
```

```python
plt.figure(figsize=(10,4))  # (widh , height ().keys())
no_of_count = list(df['brand'].value_counts().values)
plt.bar(brand_name,no_of_count,color='red',ec="black")
plt.xticks(rotation=75) 
# plt.yticks(rotation=75)
plt.savefig('brandpopulation')   # brandpopulation 
plt.show()

# ec = edge color 
```
