### barplot() 

```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
brand_name = list(df['brand'].value_counts().keys())
bike_count = list(df['brand'].value_counts().values)
```

```python
## matplotlib 
plt.bar(brand_name,bike_count)
plt.xlabel("Brands")
plt.ylabel('Bike counts')
plt.xticks(rotation=90)
plt.show()
```

```python
# seaborn 
sns.barplot(x=brand_name,y=bike_count)
plt.xlabel("Brands")
plt.ylabel('Bike counts')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x='tip',y='total_bill',data=df)
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',style='smoker')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df, hue='time',col='day')
plt.show()
```

```python
import warnings 
warnings.filterwarnings('ignore')
## to ignore the warnings 
```

```python
sns.relplot(x='size',y='total_bill',data=df,col='day',kind='line')
plt.show()
# relational plot , scatter , line  
```

## "strip", "swarm", "box", "violin", "boxen", "point", "bar", or "count".

```python
# catgorical plot 
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex')
plt.title('Strip chart')
plt.show()

# strip chart  
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='swarm')
plt.title('swarm chart')
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='box')
plt.title('swarm chart')
plt.show()
```

```python
# violin

sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='boxen')
plt.title('swarm chart')
plt.show()
```
